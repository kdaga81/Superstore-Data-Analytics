```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_excel("walmart.xls")
```

```python
dataset.head(2)
```

```python
dataset.isnull().sum()
```

```python
dataset['Discount'].value_counts().sort_index()
```

#discount

```python
sns.countplot(x="Discount", data=dataset)
plt.title("Discount Statistics")
plt.show()
```

```python
dataset['Ship Mode'].value_counts()
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values
plt.pie(y, labels= x, startangle=60, autopct="%0.2f%%")
plt.legend(loc=2)
plt.show()
```

```python
sns.countplot(x="Ship Mode", data = dataset, hue="Category")
plt.show()
```

```python
sns.countplot(x="Category", data=dataset, hue="Sub-Category")
plt.show()
```

```python
sns.countplot(x="Category", data=dataset[dataset["Category"]=="Furniture"], hue="Sub-Category")
plt.show()
```
